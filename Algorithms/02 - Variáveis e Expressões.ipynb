# **Variáveis**

São elementos fundamentais da programação e podem ser comparadas a caixas que armazenam dados na memória do computador, conforme ilustrado na Figura 1. Elas funcionam como "etiquetas" que permitem que o programador se refira aos dados de forma mais fácil e organizada, utilizando um nome significativo que represente o conteúdo da variável. Por essa razão, é essencial escolher nomes significativos para as variáveis, que possam transmitir a informação do que está sendo armazenado, facilitando a leitura e manutenção do código. É uma boa prática evitar palavras-chave ou nomes genéricos que possam gerar confusão ou dificultar a compreensão do código.

**Observação:** é importante destacar que as variáveis não são os próprios valores que elas armazenam. Elas apenas apontam para um local de memória onde os dados estão armazenados. Quando o programador faz referência a uma variável, ele está, na verdade, acessando o conteúdo armazenado naquele endereço de memória.

- Armazenar informações que serão utilizadas para tomar decisões em um programa;
- Armazenar dados que serão lidos ou gravados em um arquivo.

## **Regras para o nome de uma variável**

Para dar nomes às variáveis em *Python*, algumas regras devem ser observadas:

- Não podem conter espaços;
- Não podem iniciar com números;
- Além de caracteres alfanuméricos, pode conter somente o caracter: “_”;
- Apesar de permitido em *Python*, caracteres acentuados em nomes de variáveis devem ser evitados, pois a maioria das linguagens de programação não permite acentos em nomes de variáveis.
- *Python* diferencia letras maiúsculas e minúsculas, ou seja: Nome != nome != NOME;
- Além disso, é recomendado que as variáveis tenham nomes significativos. A escolha de nomes significativos para as variáveis ajuda o programador a entender o que o programa faz e a prevenir
erros.

Alguns exemplos de nomes inválidos em *Python*:

- 1Aluno (o primeiro caractere é um algarismo)
- total de alunos (tem espaços)
- #funcionarios (o uso do caracter inválido “#”)
- %valor (o uso do caracter inválido “%”)

## **Operador de Atribuição**

A atribuição de valores é a passagem de informação a determinada variável. Toda variável, por sua definição, pode receber valores ou então, pode ter seu valor alterado. A linguagem *Python* tem definido que o sinal que conhecemos pelo nome de igual ($ = $) será o operador de atribuição, conforme ilustrado na Figura 2.

image.png

Figura 2: Atribuição de valor para uma variável.

A parte ao lado esquerdo do operador de atribuição sempre receberá o valor definido no lado direito do operador. Sendo assim, a única estrutura de linguagem aceita do lado esquerdo do operador é um nome de variável válido. Por outro lado, qualquer expressão válida, que resulte em um valor da linguagem de programação, poderá ser usado no lado direito do operador de atribuição.

Diferente de outras linguagens de programação, em *Python* não é necessário especificar o tipo de dado na declaração de uma variável, pois a linguagem infere o tipo de uma variável no momento da primeira atribuição.

Um valor a ser atribudo à variável pode ser, por exemplo:
- valor inteiro (**int**): 2, 4, 10, etc.;
- valor real (**float**): 2.7698, 0.0023, etc.;
- valor textual (**string**): “Um texto qualquer”;
- valor lógico (**boolean**): True ou False;
- outros valores disponibilizados pela linguagem;

### **Atribuição Múltipla**

A atribuição de valores é uma tarefa corriqueira e essencial no desenvolvimento de software. Algumas linguagens como o *Python*, definem notações diferente para atribuição, permitindo assim, por exemplo, inicializar várias variáveis ou então, atribuir valores diferentes para diferentes variáveis numa mesma expressão.

A atribuição múltipla é a capacidade de atribuir um conjunto de valores a um conjunto de variáveis. Ambos conjuntos precisam, obrigatoriamente, possuir a mesma quantidade de elementos, isto é, a quantidade de variáveis a esquerda deve ter a mesma quantidade de valores a direita, como pode ser visto no código a seguir.

image.png

```python
x, y = 10, 20
temp = x
x = y
y = temp
print(f'x={x}, y={y}')
```

```python
x, y = 10, 20
y, x = x, y
print(f'x={x}, y={y}')
```

No exemplo acima, estamos declarando e atribuindo valores a 2 variáveis. A primeira variável de nome $x$, está sendo atribuido o valor 10, e a segunda variável de nome $y$, esta sendo atribuido o valor 20.

## **Operadores Relacionais**

Python possui operadores de comparação, como os da matemática: <, >, <=, >=, ==, !=. A tabela abaixo resume esses operadores de comparação. Nela assumimos que estamos fazendo $x$ `operador` $y$, ou seja, que estamos aplicando um operador a dois operandos, $x$ e $y$.

Operador |	Descrição	| Exemplo
--- | --- | ---
== | True se x e y são iguais | 5 == 2 resulta em False
!= | True se x e y são diferentes | 5 != 2 resulta em True
> | True se x é maior que y | 5 > 2 resulta em True
< | True se x é menor que y | 5 < 2 resulta em False
>= | True se x é maior ou igual a y | 5 >= 2 resulta em True
<= | True se x é menor ou igual a y | 5 <= 2 resulta em False

**Tome cuidado:** não confunda o operador $==$ (teste de igualdade) com o operador $=$ (operador de atribuição). Este último é usado para atribuir um valor a uma variável. Por exemplo: a = 20. É muito comum iniciantes confundirem esses operadores, principalmente em condicionais.

## **Operadores Lógicos**

Python nos disponibiliza três tipos de Operadores Lógicos: o **and**, o **or** e o **not**.

Operador | Definição
--- | ---
and	| Retorna True se ambas as afirmações forem verdadeiras
or	| Retorna True se pelo menos uma das afirmações for verdadeira
not	| retorna Falso se o resultado for verdadeiro

### **- and:** "e" lógico

p | operador | q | resulta
--- | --- | --- | ---
True | and | True | True
True | and | False | False
False | and | True | False
False | and  | False | False

Exemplo:

```python
p = 3 > 7 # False
q = 4 < 8 # True
print(p and q)
```

### **- or:** "ou" lógico

p | operador | q | resulta
--- | --- | --- | ---
True | or | True | True
True | or | False | True
False | or | True | True
False | or  | False | False

Exemplo:

```python
p = 3 > 7 # False
q = 4 < 8 # True
print(p or q)
```

### **- not:** "não" lógico

operador | r | resulta
--- | --- | ---
not | True | False
not | False | True

Exemplo:

```python
p = 3 > 7 # False
q = 4 < 8 # True
r = p and q # False
print(not r) # True
```

```python
p = 3 > 7 # False
q = 4 < 8 # True
r = p or q # True
print(not r) # False
```

## **Operadores de Identidade**

Estes operadores são utilizados para comparar objetos, verificando se os objetos testados referenciam o mesmo objeto (is) ou não (is not).

Operador | Definição
--- | ---
is	| Retorna True se ambas as variáveis são o mesmo objeto
is not | Retorna True se ambas as variáveis não forem o mesmo objeto

Vejamos com mais detalhes o uso dos operadores de identidade:

- Exemplo com **números**:

```python
x, y = 1, 1
```

```python
print(x is y)
```

```python
print(x is not y)
```

- Exemplo com **Strings**

```python
a, b, c = 'Sport', 'Sport', 'Santa Cruz'
```

```python
print(a is b)
```

```python
print(a is c)
```

- Exemplo com **Listas**

```python
h = [1, 2, 3]
i = [1, 2, 3]
j = h
```

```python
print(h is i)
```

```python
print(h is not j)
```

**Observação 1:** Perceba aqui que *x* e *y* são inteiros do mesmo valor, então eles são iguais e idênticos, o mesmo vale para *a* e *b* (strings). Porém, isso não ocorre com *h* e *i*, que são iguais, porém não idênticos, isso porque o interpretador localiza eles separadamente em memória.

**Observação 2:** Muitas vezes programadores ficam na dúvida em quando utilizar o operador de igualdade $==$ ou o operador de identidade *is*. Lembre-se que o operador $==$ verifica os valores testados, enquanto o operador *is* testa a referência dos valores testados!

## **Precedência de operadores**

Deve-se tomar cuidado, no entanto, ao agrupar expressões com mais de um operador. Por exemplo, observe a célula a seguir:

```python
a, b = 3, 6
c = a + b * 2
print(c)
```

É diferente de:

```python
a, b = 3, 6
c = (a + b) * 2
print(c)
```

Quando criamos uma expressão em *Python*, existe em uma ordem em que as subexpressões são avaliadas. Essa ordem é determinada por algo que chamamos de precedência de operadores. Por exemplo, se tivermos a expressão a ** 2 + b * 3 % 2, e assumirmos que a = 4 e b = 3 antes de avaliarmos a expressão, teremos a seguinte sequência de operações:

- a ** 2 + b * 3 % 2: expressão original.

- 4 ** 2 + 3 * 3 % 2: os valores de a e b são substituídos no lugar das variáveis.

- 16 + 3 * 3 % 2: a expressão de potenciação é avaliada (4 ** 2 = 16). Ela é a primeira a ser avaliada porque possui a maior precedência.

- 16 + 9 % 2: a expressão 3 * 3 é avaliada. O operador é avaliado antes do % simplesmente porque apareceu antes na expressão. Tanto o operador quanto o % possuem a mesma precedência. Quando dois operadores possuem a mesma precedência, o que aparece primeiro (mais à esquerda) na expressão é avaliado primeiro.

- 16 + 1: a expressão 9 % 2 é avaliada, pois o operador % possui maior precedência que o operador +.

- 17: a expressão 16 + 1 é avaliada

A tabela a seguir resume a precedência dos principais operadores da linguagem *Python*:

Operador | Descrição | Precedência
--- | --- | ---
or | Boolean OR | Menor
and |	Boolean AND
not	| Boolean NOT
==, !=, <, <=, >, >=, is, is not | Comparações, Identidade
\|	| Bitwise OR
^	| Bitwise XOR
&	| Bitwise AND
<<, >> | Bit Shifts
+, - | Adição, Subtração
*, /, //, %	 | Multiplicação, Divisão, Floor Division, Modulus
+x, -x, ~x	| Unário Positivo, Unário Negativo, Bitwise Negation
** | Exponenciação | Maior

Os operadores na parte superior da tabela têm a precedência mais baixa e na parte inferior da tabela têm a mais alta. Quaisquer operadores na mesma linha da tabela têm precedência igual e são avaliados da esquerda para a direita. **A exceção para isso é a exponenciação e o not**:

```python
print(2 ** 3 ** 2)
```

```python
print(not not True)
```

### **Uso de Parênteses**

É possível inserir parênteses em expressões para alterar a ordem na qual sub-expressões devem ser avaliadas. Por exemplo, a expressão 5 * 4 + 3, teremos como resultado o valor 23, pois o operador de multiplicação possui maior precedência que o operador de soma, portanto a multiplicação será realizada primeiro. Se quisermos que a soma seja realizada primeiro, podemos escrever 5 * (4 + 3), que dará 35 como resultado. Uma forma de ver isso é como se os parênteses tivessem maior precedência que todos os outros operadores, então o que está dentro deles será avaliado primeiro.

Entre os erros mais comuns que acontecem com quem está iniciando sua jornada com a linguagem Python estão os famosos SyntaxError:

```python
5 * ((4 + 3)
```

# **Glossário**

- Variável: um nome que se refere a um valor;
- Atribuição: uma instrução que atribui um valor a uma variável;
- Operando: um dos valores que um operador produz;
- Expressão: uma combinação de variáveis, operadores e valores que representa um resultado único;
- Instrução: uma seção de códigos que representa um comando ou ação;
- Modo Interativo: um modo de usar o interpretador Python, digitando o código no prompt;
- Modo Script: um modo de usar o interpretador do Python para ler códigos em um script e executá-lo;
- Script: um programa armazenado em um arquivo;
- Ordem das Operações: as regras que governam a ordem na qual as expressões que envolvem vários operadores e operandos são avaliadas;
- Comentários: informações em um programa destinadas a outros programadores que não têm efeito sobre a execução do programa;
- Erro de Sintaxe: um erro em um programa que torna sua análise impossível;
- Exceção: um erro que se descobre quando o programa é executado;
- Semântica: o significado de um programa;
- Erro Semântico: um erro que faz com que o programa faça algo diferente do que o programador pretendia.
