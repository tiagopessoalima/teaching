## **Parâmetros da função**

- São variáveis ​​declaradas no cabeçalho da função que recebem valores externos ao serem chamados. Eles são usados exclusivamente dentro do bloco de instruções da função e são necessários para que a função execute sua operação corretamente.

- Se um parâmetro não tem um valor *default* (padrão), a falta de fornecimento de um valor resultará em um erro e a função não será executada. Portanto, é importante fornecer valores para todos os parâmetros obrigatórios.

- Os parâmetros de função são uma maneira de tornar as funções mais flexíveis e genéricas, permitindo que elas possam ser usadas com diferentes valores de entrada.

### **O parâmetro `sep` da função *print***

Por padrão, a função `print` separa cada item de saída com um espaço em branco. No entanto, é possível definir um caractere de separação personalizado utilizando o parâmetro `sep`. Veja um exemplo abaixo para entender melhor:

```python
print('Dia', 'Mês', 'Ano')
```

```python
print('Dia', 'Mês', 'Ano', sep='/')
print('ontem', 'Hoje', 'Amanhã', sep='-')
print("B", "n", "n", ".", sep='a')
```

### **O parâmetro `end` da função *print***

A função `print`, por padrão, adiciona a quebra de linha '\n' como último caractere da saída. O parâmetro `end` pode ser utilizado para modificar esse comportamento, permitindo que o desenvolvedor defina qual caractere será adicionado ao final do dado impresso no terminal. Veja os exemplos a seguir para entender melhor:

```python
# Exemplo com fim de linha sem nenhum caracter
print('Vamos estudar ', end='')
print('Python')

# Exemplo com fim de linha igual à ->
print('Quando ensino', end=' -> ')
print('Aprendo')

# Exemplo com fim de linha igual à :
print("Nota", end=': ')
print(10)
```

### **Parâmetro `file` da função *print***

A função `print` pode ser utilizada para gravar dados em arquivos. Para isso, é necessário utilizar o parâmetro `file` da função `print` em conjunto com a função `open`, que permite abrir um arquivo em modo de escrita. Veja o exemplo abaixo para entender melhor:

```python
with open('arquivo.txt', 'w') as arquivo:
    print('Primeira linha do arquivo', file=arquivo)
    print('Segunda linha do arquivo', file=arquivo)
```

- A função `open` abre arquivos. Ela recebe dois parâmetros: o primeiro é o nome do arquivo a ser aberto e o segundo é o modo que queremos trabalhar com esse arquivo - se queremos ler ou escrever. O modo é passado como uma string: 'w' (abreviação para write) para escrita, 'r' (abreviação para read) para leitura e 'a' (abreviação para appending) para adicionar conteúdo ao final do arquivo.

- A instrução *with* é utilizada para garantir que os recursos adquiridos sejam finalizados corretamente, como o arquivo aberto. No exemplo acima, utilizamos a instrução *with* para abrir o arquivo, escrever nele e encerrá-lo de forma segura.

Embora a instrução *with* seja útil para garantir que recursos adquiridos sejam finalizados corretamente, ela não lida com exceções que podem ocorrer durante a execução do bloco de código dentro do *with*. Para lidar com [exceções](https://www.youtube.com/watch?v=xz2B3bfNjEk), podemos usar a estrutura *try-except-finally*.

- A estrutura *try-except* tenta executar um bloco de código e, caso uma exceção ocorra, o bloco de código dentro do *except* é executado. Já a cláusula *finally* é executada independentemente de ter ocorrido ou não uma exceção.

Por exemplo, se quisermos abrir um arquivo e escrever nele, podemos usar a seguinte estrutura:

```python
try:
  arquivo = open('arquivo.txt', 'r')
  print('Primeira linha do arquivo', file=arquivo)
  print('Segunda linha do arquivo', file=arquivo)
except IOError:
  print('Ocorreu um erro ao tentar escrever no arquivo.')
finally:
  print('fechou!')
  arquivo.close()
```

Neste exemplo, a cláusula *try* tenta abrir o arquivo e escrever nele. Caso ocorra um erro de leitura ou escrita (*IOError*), o bloco de código dentro do except é executado, exibindo uma mensagem de erro. A cláusula *finally* garante que o arquivo seja fechado, independentemente de ocorrer ou não uma exceção durante a execução do bloco *try*.

3. Escreva um código que recebe uma lista de números e apresenta o desvio padrão dos mesmos.

Passos:

- Solicitar uma lista de números.
- Calcular a média dos valores na lista.
- Para cada número na lista, calcular o quadrado da diferença entre o número e a média.
- Somar todos os quadrados calculados no passo anterior.
- Dividir o resultado da soma pelo número total de elementos na lista.
- Calcular a raiz quadrada do resultado da divisão do passo anterior.
- Exibir o resultado usando a função `print`.

Sem compreensão de listas

```python
lista = [1,2,3,4,5,6,7,8,9]

media = sum(lista) / len(lista)

soma = 0
for numero in lista:
  diferenca = numero - media
  quadrado = math.pow(diferenca, 2)
  soma += quadrado

divisao = soma / len(lista)

desvio = math.sqrt(divisao)

print(desvio)
```

Com compreensão de listas

```python
lista = [1,2,3,4,5,6,7,8,9]

media = sum(lista) / len(lista)

quadrados = [math.pow(numero-media,2) for numero in lista]

soma = sum(quadrados)

divisao = soma / len(lista)

desvio = math.sqrt(divisao)

print(desvio)
```

Para criar a função, teríamos:

```python
def desvio_padrao(lista):
  media = sum(lista) / len(lista)
  quadrados = [math.pow(numero-media,2) for numero in lista]
  soma = sum(quadrados)
  divisao = soma / len(lista)
  desvio = math.sqrt(divisao)
  print(desvio)
```

```python
lista = [1,2,3,4,5,6,7,8,9]
desvio_padrao(lista)
```

## **Fluxo de Execução**

Para garantir que uma função seja definida antes do seu primeiro uso, é preciso saber a ordem na qual as instruções serão executadas. A execução sempre inicia na primeira instrução do programa. As instruções são executadas uma após a outra, de cima para baixo.

As definições de função não alteram o fluxo de execução do programa, mas lembre-se de que as instruções dentro da função não são executadas até que a função seja chamada. Uma chamada de função é um desvio no fluxo de execução. Em vez de ir para a instrução seguinte, o fluxo salta para o corpo da função, executa as instruções lá, e então volta para continuar de onde parou.

## **Parâmetros e Argumentos**

Algumas funções que vimos até aqui exigem argumentos. Por exemplo, ao chamar *math.sin*, você usa um número como argumento. Algumas funções exigem mais de um argumento: o *math.pow* exige dois, a base e o expoente. Dentro da função, os argumentos são atribuídos a variáveis chamada **parâmetros**. ## **Passagem por Valor**

Quando se passa uma variável a uma função, o *Python* passa a referência ao objeto ao qual a variável se refere. Assim, se o valor é imutável, a função não modifica a variável chamada. Se o valor é mutável, a função pode modificar a variável chamada. Observe:

```python
def modificar(x, y, z):
  x+=1
  y.append(42)
  z = [99] # nova referencia
  print('x',x)
  print('y',y)
  print('z',z)

a = 77   # variável imutável
b = [99] # variável mutável
c = [28]
modificar(a, b, c)

print('a', a)
print('b', b)
print('c', c)
```

Este código define uma função chamada modificar que recebe três parâmetros: x, y e z. A função modifica o valor de x adicionando 1, adiciona o número 42 à lista y e cria uma nova referência para a lista z, atribuindo-a a uma nova lista contendo apenas o número 99. Depois disso, a função imprime os valores atualizados de x, y e z.

O código então define as variáveis a, b e c. A variável a é um número imutável e a variável b é uma lista mutável contendo o número 99. A variável c também é uma lista contendo o número 28.

Em seguida, a função modificar é chamada com a, b e c como argumentos. Depois disso, o código imprime os valores atualizados de a, b e c.

Podemos observar que x é incrementado em 1, y recebe o número 42 adicionado ao final, e z recebe uma nova referência para uma lista contendo apenas o número 99. No entanto, a variável a (que é imutável) não é afetada pela função modificar. Já a lista b (que é mutável) é alterada, recebendo o número 42 adicionado ao final, e a lista c continua a mesma, já que a variável z recebeu uma nova referência e não a alterou.

É importante notar que quando a função modificar cria uma nova referência para z, isso não afeta a variável c fora da função, já que a nova referência é atribuída apenas dentro da função.

## **Número de parâmetros variáveis**


É possível especifica um função que pode receber uma quantidade indefinida de argumentos. Esses parâmetros serão "empacotados" em uma tupla chamada args. No código abaixo, note que é preciso marcar a tupla *args* com um asterisco, para que *Python* saiba que deve alocar todos os parâmetros como elementos dela, na ordem que são passados.

```python
def func_args(*args):
  print(f'tipo: {type(args)} conteúdo: {args}')
  for arg in args:
    print(f'tipo: {type(arg)} conteúdo: {arg}')

func_args(1, 'A', {'valor': 10})
```

*Python* também permite que uma função seja chamada com uma quantidade indefinida de parâmetros por nome. Para isso, é preciso declarar o parâmetro *kwargs*, que irá representar um dicionário em que *Python* irá alocar todos os parâmetros passados por nome que não foram pré-definidos na chamada da função. O parâmetro *kwargs* deve ser marcado por asteriscos duplos na definição da função. Exemplo:

```python
def func_kwargs(**kwargs):
    print(f'tipo: {type(kwargs)} conteúdo: {kwargs}')
    for key, value in kwargs.items():
        print(f'atributo: {key}, valor: {value}')

func_kwargs(nome='James', sobrenome='Bond', cargo='Agente 007')
```

**Observação:** As palavras args e kwargs são usadas como convenção, ou seja, poderiam ter qualquer outro nome, por exemplo: \*tiago e **lima.

## **Funções com Resultado e Funções Nulas**

Diversas funções que já usamos, como as funções matemáticas, devolvem resultado. Outras funções, como print, executam uma ação, mas não devolvem um valor. Se você atribuir o resultado a uma variável, recebe um valor especial chamado **None**:

```python
resultado = print_2_vezes(':-)')
```

```python
print(resultado)
```

O valor **None** não é o mesmo que a string 'None'. É um valor especial que tem seu próprio tipo:

```python
print(type(resultado))
```

As funções que criamos até aqui são nulas, ou seja, não devolvem um valor. Agora vejamos uma função que não possui um comando **return**:

```python
def potencia(x, y):
  return x ** y

resultado = potencia(2,3)

print(resultado)
```

Também é possível construir funções que retornam múltiplos valores em *Python*. Vejamos uma função que recebe dois inteiros e retorna o quociente e o resto da sua divisão:

```python
def div(dividendo, divisor):
  return dividendo // divisor, dividendo % divisor

resultado = div(2,3)

print(resultado)
```

```python
print(type(resultado))
```

Como se pode ver, os retornos da função `div` foram impressos entre parênteses, ou seja, na forma de uma tupla. Isso significa que Python emula retornos múltiplos de funções por meio do retorno de um único valor do tipo tupla. Essa abordagem permite desconstruir o retorno em múltiplas atribuições durante a chamada. Exemplo:

```python
dividendo, divisor = 2, 3
quociente, resto = div(dividendo, divisor)
print(quociente, resto)
```

## ***yield***

Observe as duas célula abaixo:

```python
list_compr = [x for x in range(10)]
print(list_compr)
```

```python
list_gener = (x for x in range(10))
print(list_gener)
```

Perceba que a primeira forma retorna a lista normalmente, usando *list* comprehension de modo padrão *Python*. A segunda retorna um objeto da classe *generator*, que "gera" os valores sob demanda. Para "gerar" os valores do `list_gener` podemos utilizar:

- Função `next`: para gerar o próximo elemento;

```python
print(next(list_gener))
```

- `for`: para ir gerando nas iterações do loop.

```python
for i in list_gener:
  print(i)
```

A coisa começa a ficar interessante quando estamos trabalhando com funções. Imagine que estamos escrevendo uma função para retornar a lista dos quadrados dos números de `x` até `y`. No entanto, gostaríamos de receber um elemento da lista, executar operações, depois gerar outro elemento, executar operações, e assim por diante. Veja como ficaria isso:

```python
def operacoes(number):
  print('faça operações com:', number)

def gera_quadrados_no_intervalo(x, y):
  for i in range(x, y):
    yield i*i

x, y = 10, 20

for numero in gera_quadrados_no_intervalo(x, y):
  operacoes(numero)
```

Perceba, o *yield* nesse código funciona mais ou menos como um *return*, entretanto, mantém salvo o contexto da função, e quando o `for` fora da função itera novamente, a função `gera_quadrados_no_intervalo` volta a ser chamada, continuando sua execução de onde parou, mantendo o contexto da sua iteração, seguindo a sequência correta do `range(x, y)`.

Vamos agora trocar o *yield* por *return* (e alterar o loop fora da função para ter uma lista). Nesse caso o resultado será sempre 100, pois o return não salva o contexto da função, executando-a "do zero" toda vez. Veja abaixo:

```python
def operacoes(number):
  print('faça operações com:', number)

def gera_quadrados_no_intervalo(x, y):
  for i in range(x, y):
    return i*i

x, y = 10, 20

for _ in range(x, y):
  numero = gera_quadrados_no_intervalo(x, y)
  operacoes(numero)
```

Mas é possível simplesmente carregar todos os valores em uma lista e iterá-los normalmente no for. Qual vantagem do *yield*?

Pense o seguinte: e se essa lista tiver 1 Milhão de itens e você continuar sem saber direito quanto tempo os cálculos vão demorar, você carregaria os 1 Milhão de itens na memória e deixaria lá até os cálculos terminarem?

Vamos usar a função `getsizeof` do `sys` para medir essa diferença. Ela retorna o tamanho em *bytes* do objeto passado por parâmetro.

```python
from sys import getsizeof

um_milhao = 10**6

list_compr = [x for x in range(um_milhao)]
print(getsizeof(list_compr))

list_gener = (x for x in range(um_milhao))
print(getsizeof(list_gener))
```

Veja que uma lista com o range de 1 Milhão utiliza 8697464 de *bytes*, enquanto generator utiliza apenas 128 *bytes*. Dividindo um pelo outro (8697464/128), você chega em uma razão de aproximadamente de 68000 vezes!

# **Glossário**
- Função: uma sequência nomeada de declarações que executa alguma operação. Podem receber argumentos ou produzir um resultado;
- Definição de Função: uma instrução que cria uma função nova, especificando seu nome, parâmetros e as instruções;
- Objeto da Função: um valor é criado por uma definição de função. O nome da função é uma variável que se refere a um objeto de função;
- Cabeçalho: a primeira linha de uma função;
- Corpo: a sequência de instruções dentro de uma função
- Parâmetro: um nome usado dentro de uma função para se referir ao valor passado como argumento;
- Chamada de Função: instrução que executa uma função. É composta pelo nome da função seguido de argumentos entre parênteses;
- Variável Local: variável definida dentro de uma função e por este motivo só pode ser visualizada dentro da função;
- Valor de Retorno: o resultado de uma função;
- Função com Resultado: uma função que devolve um valor;
- Função Nula: uma função que devolve None;
- None: um valor especial apresentado por funções nulas;
- Módulo: um arquivo que contém uma coleção de funções relacionadas e outras definições;
- Instrução de Importação: uma instrução que lê um módulo e cria um objeto de módulo;
- Objeto de Módulo: um valor criado por uma instrução import que oferece acesso aos valores definidos em um módulo;
- Notação de Ponto: sintaxe para chamar funções em outro módulo - o nome do módulo seguido do ponto e o nome da função;
- Fluxo de Execução: a ordem na qual as instruções são executadas.
