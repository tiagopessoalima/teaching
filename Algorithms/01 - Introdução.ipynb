# **Como funciona um computador?**

Um computador funciona baseado em três principais elementos: Processador, Memória e Armazenamento. Mais detalhes a respeito serão vistos nas disciplinas de Arquitetura e Organização de Computadores. O básico para programação está neste [vídeo](https://youtu.be/MpKbTNonIwc).

# **O que é um programa?**



Um programa é uma sequência de instruções que especifica como executar uma operação de computação. Ex: a operação de computação pode ser algo matemático, como solucionar um sistema de equações ou encontrar as raízes de um polinômio, mas também pode ser uma operação de computação simbólica, como a busca e a substituição de textos em um documento; ou algo gráfico, como o processamento de uma imagem ou a reprodução de um vídeo.

# **Instruções básicas**

- **Entrada**: receber dados do teclado, de um arquivo, da rede ou de algum outro dispositivo.

- **Processamento**: executar operações matemáticas básicas como adição e multiplicação.

  - **Execução condicional**: verificar a existência de certas condições e executar o código adequado.

  - **Repetição**: executar várias vezes alguma ação, normalmente com algumas variações.

- **Saída**: exibir dados na tela, salvá-los em um arquivo, enviá-los pela rede etc.

Por questões de simplicidade, utilizaremos em nosso primeiro algoritmos em 3 instruções: Entrada, Processamento e Saída. A célula abaixo mostra o código em *Python*.

# **Programação Estruturada**

Os fluxogramas são uma forma de representar um programa de forma mais “alto nível” para entendermos melhor a lógica dos programas implementados. ## **Exemplo 1**

Suponha que na aula de física você aprendeu a converter uma temperatura de Celsius ($TC$) para Fahrenheit ($TF$). A fórmula usada é:

$TF = (TC*9/5)+32$

Em uma sequência sistemática de passos, o raciocínio pode ser o seguinte:

- **Entrada**:
  - Ler um valor de temperatura em Celsius;
- **Processamento**
  - Multiplicar TC por 9 e dividir por 5;
  - Somar 32 ao valor encontrado;
- **Saída**:
  - Imprimir o valor da saída.

Vamos agora organizar esse programa em um fluxograma, para visualizar os passos que o programa vai tomar. Observe na Figura 4, que o fluxograma apenas organiza os passos antes representados por uma lista textual.

image.png

Figura 4: Fluxograma para o Exemplo 1.

O fluxograma mostra a sequência de passos para atingir o objetivo. Abaixo, podemos ver o funcionamento disso em um programa em *Python*.

```python
# Entrada
C = float(input('Digite a temperatura em Celsius: '))

# Processamento
F = C * 9 / 5
F = F + 32

#Saída
print(f'A temperatura em Fahrenheit é {F:.1f}')
```

Observe que para este exemplo foram apresentadas 3 representações para se resolver algoritmicamente o problema. Primeiro de forma textual, depois em um fluxograma e, por fim, em um programa em *Python*. Todas estas representações estão corretas e resolvem o problema, porém, quando utilizamos uma linguagem de programação é possível executar o programa e ver o resultado final.
