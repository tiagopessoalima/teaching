### shape

Esse atributo indica a forma, por exemplo:

```python
print(v.shape)
```

Neste caso, o `ndarray` `v` possui `1` dimensão (ou eixo) com `4` elementos. Um `ndarray` unidimensional corresponde a um vetor. Podemos também criar um `ndarray` bidimensional (que corresponde a uma matriz) usando o atributo `shape`:

```python
v = np.array([1,2,3,4])
v.shape = (2,2)
print('v =\n',v)
```

Outra forma útil de mudar o `shape` de um `ndarray` é simplesmente utilizando a função `reshape`:

```python
v = np.array([1,2,3,4])
print(v.reshape(2,2))
```

A função `reshape` aceita o uso de `-1` como dimensão na especificação do `shape`. Neste caso, a função `reshape` determinará o número de elementos em cada dimensão de forma a manter inalterado o número total de elementos no `ndarray`. Exemplo:

```python
v = np.array([1,2,3,4,5,6,7,8])
print(v.reshape(2,-1)) # print(v.reshape(3,-1)) ?
```

### ndim e size

O número de eixos (ou dimensões) de um `ndarray` é dado pelo atributo `ndim`, enquanto o número total de elementos é dado por `size`:

```python
v = np.array(range(24)).reshape(2,3,4)
print('shape =', v.shape)
print('dimensões =', v.ndim) # mesmo que len(v.shape)
print('elementos =', v.size) 
print('v =\n', v)
```

### Resumindo

#### Um `ndarray` de `1` dimensão é um objeto similar a um vetor (`rank = 1`): 

- `arr1D = ([a0,a1,...,aN-1])`, `arr1D.size = N`, `arr1D.ndim = 1`, `arr1D.shape = (N,)`.

#### Um `ndarray` de `2` dimensões é um objeto similar a uma matriz (`rank = 2`): 

- se possui `M` linhas, cada uma com `N` elementos, então `arr2D.size = M × N`, `arr2D.ndim = 2`, `arr2D.shape = (M,N)`.

#### Um `ndarray` de `3` dimensões é uma coleção de matrizes (`rank = 3`): 

- se possui `K` matrizes de `M` linhas, cada uma com `N` elementos, então `arr3D.size = K × M × N` , `arr3D.ndim = 3`, `arr3D.shape = (K, M, N)`.

#### `Ndarrays` de ordem superior são generalizações desse processo, acrescentados novos eixos.

## Cópias e Vistas

Ao operar com `ndarray`, os dados são às vezes copiados em um novo `ndarray` e às vezes não, o que pode confundir iniciantes. Por exemplo, atribuições simples não criam cópias do `array`, ou seja, elas apenas criam referências/ponteiros para o mesmo objeto:

```python
a = np.arange(12)
b = a
print(b is a)
print('a =\n', a)
```

```python
b.shape = 3,4 # outra forma de mudar a forma de um array
print('shape =',a.shape) # o comando acima muda a forma de a também
print('a =\n', a)
```

Ao chamar funções, *Python* passa objetos mutáveis como referências, portanto quando um array é passado como argumento para uma função, o mesmo objeto estará disponível no escopo da função chamada:

```python
def f(x):
  print(id(x))

print(id(a))
f(a)
```

O método `view` cria um novo objeto, mas esse novo objeto “olha” para os mesmos dados, ou seja, ele é uma “vista”. Observe:

```python
c = a.view()
print(c is a)
print(c.base is a)
```

```python
c.shape = 2,6
print(a.shape) # a não muda de forma
```

```python
c[0,4] = 1234 # muda em a também
print('a =\n',a)
```

```python
print('c =\n',c)
```

Uma fatia de um `ndarray` retorna uma **vista**:

```python
s = a[:,1:3]
s[:] = 10
print('a =\n',a)
```

Para obter uma cópia independente de um `ndarray`, usa-se o método `copy`:

```python
d = a.copy()
print(d is a)
print(d.base is a) # d não compartilha dados com a
```

```python
d[0,0] = 9999
print('a =\n',a)
```

Em `ndarray` com *rank* maior ou igual a `2` pode-se identificar o eixo em que se deseja aplicar as funções `min`, `max`, etc. Caso o eixo não seja especificado, a funções retornará o menor/maior valor entre todos incluindo todos os eixos:

Um erro muito comum no uso do `NumPy` é não explorar sua eficiência computacional por meio das operações vetoriais. Dentro do possível, deve-se sempre organizar implementação de um algoritmo utilizando somente operações sobre `ndarrays`. Por exemplo, abaixo é feita a comparação entre o tempo computacional em usar a operação dot e uma implementação da mesma operação utilizando loops:

```python
def produto_interno(u, v):
  prod = 0
  for i in range(u.size):
    prod+=u[i]*v[i]
  return prod

u = np.random.rand(10000)
v = np.random.rand(10000)

%time _ = produto_interno(u,v)
%time _ = np.dot(u,v)
```

*Wall time* é o tempo gasto pelo código para ser executado. *CPU time* é uma medida do tempo gasto pelo processador apenas quando esteve operando sobre a tarefa. Dependendo do cálculo, o *NumPy* pode ser mais de `100` vezes mais rápido que uma operação similar em *Python*. Isso ocorre porque a operação `dot` é implementada em linguagem C e realiza a operação de maneira vetorizada.

### Operações sobre matrizes

#### Multiplicação de vetores e matrizes é realizada por meio da função `dot`:

A multiplicação de matrizes é feita por meio de um algoritmo que exige bastante atenção. Para que exista o produto entre a matriz `A` e a matriz `B`, é necessário que o número de colunas da primeira matriz, no caso `A`, seja igual ao número de linhas da segunda matriz, no caso `B`.

```python
A = np.array([[2, 3], [4, 6]])
B = np.array([[1, 3, 0], [2, 1, 1]])
print('A =\n',A)
print('B =\n',B)
print('A.B =\n',np.dot(A,B)) # A * B
```

## Operadores lógicos

As operações lógicas entre `ndarray` ocorrem elemento a elemento. O resultado é um `ndarray` booleano com valores True ou False:

```python
u = np.arange(4).reshape(2,2)
print('u =\n',u)
```

```python
v = 2*np.ones((2,2))
print('v =\n',v)
```

```python
w = u > v
print('w =\n',w)
```

Os `ndarrays` booleanos podem ser usados para filtrar valores em um `ndarray`, por exemplo:

```python
u = np.array([-1, 2, -3])
v = np.array([True, False, True])
print(u[v])
```

Note que podemos obter o mesmo efeito anterior de filtrar somente os valores negativos fazendo:

```python
w = u[u < 0]
print('w =\n',w)
```

É possível atribuir um valor aos elementos de um `ndarray` que satisfazem uma certa condições. Por exemplo, zerar os valores negativos:

```python
print('u =\n', u)
u[u < 0] = 0
print('u =\n', u)
```

As funções `all` e `any` são muito úteis em comparações lógicas.

- `all` retorna `True` se todos os elementos de um `ndarray` booleano forem `True`:

```python
v = np.arange(5)
u = v >= 0
print('u =', u)
print(np.all(u))
```

- `any` retorna `True` se pelo menos um elemento de um `ndarray` booleano for `True`:

```python
v = np.array([-1, 1, 2, 3])
u = v > 0
print('u =\n',u)
print(np.any(u))
```

## Ordenando `arrays`

NumPy oferece funções para ordenar arrays. A principal é `sort`, que retorna uma cópia ordenada do `array` passado como parâmetro:

```python
a = np.array([12, 9, 13, -1, 0, -4, 2, 7, 1])
print(np.sort(a))
```

Caso a seja um array multidimensional, pode-se usar o parâmetro axis para indicar através de qual eixo a ordenação deve ser feita.

```python
b = np.array([[12, 9, 13], [-1, 0, -4], [2, 7, 1]])
```

```python
print(np.sort(b, axis=None))  # se axis for None, o array é aplainado antes de ordenar
```

```python
print(np.sort(b, axis=0))  # axis=0 irá ordenar através das linhas
```

```python
print(np.sort(b, axis=1))  # axis=1 irá ordenar através das colunas
```

Também é possível escolher o algoritmo de ordenação usando o parâmetro `kind` (cujo valor padrão é 'quicksort'):

```python
print(np.sort(b, kind='mergesort'))
```

Outra função importante é `argsort`, que ao invés de retornar o `array` ordenado, retorna os índices do `array` nas posições corretas para ordená-lo. A função `argsort` aceita os mesmos parâmetros da função `sort`.

```python
print(np.argsort(b))
```

```python
print(a[np.argsort(b)])
```

Os dois principais objetos existentes no pandas são as séries (*Series*) e as tabelas (*DataFrame*).

### Series

Abaixo **criamos uma *Series*** por meio de uma lista de valores, incluindo o valor `NaN` (*not a number*), que equivale a uma posição nula. O *pandas* criou um índice númerico para cada linha e determinou o tipo de que melhor se representaria os dados (*float64*).

Abaixo **criamos um *DataFrame*** composto por uma matriz de inteiros. Após a chamada ao construtor `pd.DataFrame`, *pandas* cria um índice númerico para cada linha e atribui um número como título de cada coluna.

## Combinando *Series* e *DataFrames*

A função `concat` faz o trabalho de concatenação ao longo de um eixo, usando lógica de conjuntos (união ou interseção) dos índices ou colunas no outro eixo, se o outro eixo existir (Series tem apenas um eixo). Exemplo:

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
 
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
 
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D8', 'D9', 'D10', 'D11']},
                    index=[8, 9, 10, 11])


frames = [df1, df2, df3]
result = pd.concat(frames)
result
```

Supondo que você quer associar chaves específicas para os dados que pertencem a cada *DataFrame*. Para isso, usar o argumento `keys`:

```python
result = pd.concat(frames, keys=['x', 'y', 'y'])
result
```

O mesmo efeito pode ser obtido passando um dicionário com os *DataFrames* parciais:



```python
pieces = {'x': df1, 'y': df2, 'z': df3}

result = pd.concat(pieces)

result
```

O resultado passou a ter um índice múltiplo e hierárquico, o que permite selecionar cada bloco original usando o atributo `loc`. 

```python
result.loc['x']
```

```python
result.loc['y']
```

```python
result.loc['z']
```

Ao juntar múltiplos *DataFrames*, é possível escolher como lidar com os outros eixos de duas formas usando o argumento `join`:

- Tomando a união, ou seja, `join='outer'`. 

- Tomando a interseção, ou seja, `join='inner'`.

Primeiro vejamos um exemplo de `join='outer'`:

```python
df1
```

```python
df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                    index=[2, 3, 6, 7])
df4
```

```python
result = pd.concat([df1, df4], axis=0, join='outer')
result
```

```python
result = pd.concat([df1, df4], axis=1, join='outer')
result
```

Agora vejamos um exemplo de `join='inner'`:

```python
result = pd.concat([df1, df4], axis=0, join='inner')
result
```

```python
result = pd.concat([df1, df4], axis=1, join='inner')
result
```

É possível concatenar uma mistura de objetos *Series* e *DataFrame*. Internamente, *pandas* transforma a(s) *Series* em *DataFrame*(s) com apenas uma coluna com o(s) mesmo(s) nome(s) da(s) Series.

```python
s1 = pd.Series(['X0', 'X1', 'X2', 'X3'], name='X')

pd.concat([df1, s1], axis=1)
```

Caso a(s) *Series* não tenha(m) nome, o(s) nome(s) será(ão) atribuído(s) numericamente e consecutivamente:

```python
s2 = pd.Series(['_0', '_1', '_2', '_3'])

pd.concat([df1, s2, s2, s2], axis=1)
```

## Iterando *Series* e *DataFrames*

Ao iterar sobre uma *Series*, o comportamento é igual a um *array* unidimensional, produzindo os seus valores.

```python
print(df1)
```

Para iterar sobre as linhas de um *DataFrame*, *pandas* oferece os seguintes métodos: `iterrows()` e `itertuples()`.

```python
df = pd.DataFrame(
    {'a': [1, 2, 3],
     'b': ['a', 'b', 'c']}
)

print(df)
```

```python
for row_index, row in df.iterrows():
  print(row_index, row, sep='\n')
```

- `itertuples()` retorna uma tupla nomeada para cada linha. 

```python
for row in df.itertuples():
  print(row)
```

## Dividindo um objeto em grupos

Objetos *pandas* podem ser divididos através de qualquer de seus eixos por meio de um mapeamento de valores para nomes de grupos.

```python
df = pd.DataFrame([('bird', 'Falconiformes', 389.0),
                   ('bird', 'Psittaciformes', 24.0),
                   ('mammal', 'Carnivora', 80.2),
                   ('mammal', 'Primates', np.nan),
                   ('mammal', 'Carnivora', 58)],
                  index=['falcon', 'parrot', 'lion', 'monkey', 'leopard'],
                  columns=('class', 'order', 'max_speed'))

df
```

```python
grouped = df.groupby('class')
```

O agrupamento cria um objeto do tipo *DataFrameGroupBy*.

```python
print(type(grouped))
```

Note que objetos do tipo DataFrameGroupBy não têm comportamento definido de impressão. 

```python
print(grouped)
```

Eles existem para facilitar operações agrupadas. No entanto, é possível iterar sobre os diferentes grupos:

```python
for name_of_the_group, group in grouped:
  print(name_of_the_group)
  print(group)
  print()
```
