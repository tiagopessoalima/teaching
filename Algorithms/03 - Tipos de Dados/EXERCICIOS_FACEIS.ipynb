C) Escreva um código que imprima a quantidade de caracteres da string.

```python
len(texto)
```

E se eu quisesse contar apenas as letras do alfabeto?

```python
len(texto) - texto.count(' ') # Elimino os espaços!
```

3. Considere a seguinte lista de tuplas, cada uma representando um produto e
seu respectivo preço:

```python
produtos = [("Camisa", 29.99), ("Calça", 59.99), ("Meia", 9.99), ("Tênis", 99.99), ("Boné",
19.99)]
```

A) Escreva um código que imprima apenas os nomes dos produtos, sem os preços.

```python
for produto in produtos:
  print(produto[0])
```

B) Escreva um código que imprima apenas os preços dos produtos, sem os nomes.

```python
for produto in produtos:
  print(produto[1])
```

C) Escreva um código que calcule o preço médio dos produtos.

```python
soma = 0
for produto in produtos:
  soma += produto[1]
print(soma/len(produtos))
```

```python
precos = []
for produto in produtos:
  precos.append(produto[1])
print(sum(precos)/len(precos))
```

D) Escreva um código que encontre o produto mais caro da lista e imprima seu nome e
preço.

```python
max(precos)
```

```python
max(produtos)
```
