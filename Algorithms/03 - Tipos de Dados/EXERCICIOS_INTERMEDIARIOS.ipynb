a) Escreva um código que imprima todas as palavras da string em ordem alfabética,
sem repetição.

```python
texto = "Um passarinho me contou que viu um gato de botas tomando um café com um pinguim".lower()

palavras = texto.split(' ')

conjunto = sorted(set(palavras))

print(conjunto)
```

b) Escreva um código que conte quantas vezes a letra "a" aparece no texto (não
diferencia maiúsculas de minúsculas).

```python
texto = "Um passarinho me contou que viu um gato de botas tomando um café com um pinguim"

print(texto.count('a'))
```

c) Escreva um código que imprima todas as palavras do texto em ordem alfabética,
com a primeira letra em maiúscula.

```python
texto = "Um passarinho me contou que viu um gato de botas tomando um café com um pinguim".lower()

palavras = texto.split(' ')

palavras = [palavra.capitalize() for palavra in sorted(set(palavras))]

print(palavras)
```

```python
texto = "Um passarinho me contou que viu um gato de botas tomando um café com um pinguim".lower()

palavras = texto.split(' ')

palavras = sorted(set(palavras))

maiusculas = []

for palavra in palavras:
  maiusculas.append(palavra.capitalize())

print(maiusculas)
```

d) Escreva um código que imprima todas as palavras do texto em ordem decrescente
de tamanho.

```python
texto = "Um passarinho me contou que viu um gato de botas tomando um café com um pinguim".lower()

lista = texto.split(' ')

ordem = sorted(lista,key=len,reverse=True)

print(ordem)
```
