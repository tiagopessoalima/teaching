## **Leis da Recursividade**

- Um algoritmo recursivo deve **possuir um caso base** (*base case*).
- Um algoritmo recursivo deve modificar o seu estado e **se aproximar do caso base**.
- Um algoritmo recursivo **deve chamar a si mesmo**, recursivamente.

## **Sequncia de Fibonacci**

Definiremos agora uma sequncia matemtica muito famosa chamada Sequncia de Fibonacci: 1,1,2,3,5,8,13,. Nessa sequncia, os dois termos iniciais so 0 e 1, e cada termo subsequente  igual  soma dos dois termos anteriores. O clculo do n-simo termo da sequncia de Fibonacci pode ser definido recursivamente da seguinte forma:

\begin{align*}
	F_n&=
	\begin{cases}
		1 & \mbox{ se } n=0 \mbox{ ou } n=1,\\
		F_{n-1}+F_{n-2}& \mbox{ caso contrrio }.
	\end{cases}
\end{align*}

A partir da definio acima, podemos implementar o clculo do n-simo termo da sequncia assim:

```python
def fib(n):
  if n<=1: # Caso base: clculo dos dois primeiros termos.
    return 1
  else:
    return fib(n-1) + fib(n-2) # Clculo do n-simo termo.

%time fib(70)
```

Novamente, note a correspondncia direta entre a definio e sua implementao. Entretanto, a implementao acima  ineficiente porque ela faz uma grande quantidade de trabalho repetido. Para ilustrar isso, veja a figura abaixo, ilustrando o clculo de fib(5), que mostra as muitas chamadas repetidas da funo `fib`, que, por questes de espao, escrevemos simplesmente como `f`.

image.png

Existem formas mais eficientes de implementarmos a funo `fib`, sem usar recurso, como a seguinte:

```python
def fib(n):
  if n <= 1:
    return n
  fib_n_menos_2 = 0
  fib_n_menos_1 = 1
  for i in range(2, n):
    fib_n = fib_n_menos_1 + fib_n_menos_2
    fib_n_menos_2 = fib_n_menos_1
    fib_n_menos_1 = fib_n
  return fib_n_menos_1 + fib_n_menos_2

%time fib(70)
```

Pelo exemplo da funo `fib` recursiva (e sua respectiva verso no recursiva), voc pode pensar que recurso no  um recurso to vantajoso assim. Afinal, a verso iterativa no  muito mais complicada que a verso recursiva correspondente, e a verso iterativa  mais eficiente. Na prtica, a verso no recursiva de um algoritmo tende a ser mais eficiente que a verso recursiva. Entretanto, em alguns casos a verso no recursiva  to mais complicada que acaba sendo mais vantajoso usar a verso recursiva (por questes de legibilidade e facilidade de manuteno). Em outros casos (como o nosso prximo exemplo), a prpria natureza recursiva do problema e a elevada complexidade computacional inerente a ele (uma soluo iterativa no traria ganho significativo de desempenho) fazem com que recurso seja a melhor escolha.

## **Torres de Hanoi**

A lenda associada a esse problema  a seguinte: no templo de Brahma em Benares, debaixo do domo que marca o centro do mundo, existem 64 discos de ouro puro que os monges carregam, um de cada vez, para trs agulhas de diamante. Eles fazem isso em obedincia  ordenana imutvel do deus Brahma, segundo a qual nenhum disco pode ser colocado em cima de um disco menor que ele.

image.png

Segundo essa lenda, no comeo do mundo, os `64` discos formavam uma torre (empilhados do maior para o menor) em uma das agulhas de diamante. O deus Brahma ento ordenou aos monges que transferissem a torre de discos de uma agulha para outra, movendo um disco de cada vez e usando como "agulha auxiliar" a terceira agulha, e lhes disse que quando eles terminarem essa tarefa, chegar o fim do mundo.

```python
def hanoi(n, origem, destino, aux):
  if n == 1: # Caso base: movendo um disco.
    print("Mova de {} para {}".format(origem, destino))
  else:
    hanoi(n-1,origem,aux,destino) # Movendo n-1 discos de 'origem' para 'aux'.
    hanoi(1,origem,destino,aux)   # Movendo um nico disco de 'origem' para 'destino'.
    hanoi(n-1,aux,destino,origem) # Movendo os n-1 discos de 'aux' para 'destino'.

%time hanoi(64, 'A', 'C', 'B')
```

## **Calculando o Fatorial**




Considere-se o seguinte exemplo, que implementa a funo fatorial. Recorde-se que para um nmero natural $n$ se tem:

\begin{align*}
	n!&=
	\begin{cases}
		1 & \mbox{ se } n=0,\\
		n.(n-1)!& \mbox{ caso contrrio }.
	\end{cases}
\end{align*}

Ento o fatorial de 4  calculado por:

4 x 3 x 2 x 1 = 24

O mdulo math, que vimos anteriormente, j disponibiliza uma implementao.

```python
from math import factorial as fact
print(fact(4))
```

Este fato no deve impedir-nos de programarmos a nossa prpria definio. Isto  algo que faremos muitas vezes, no por razes de eficincia (a nossa definio ser sempre menos eficiente que uma definio hiper-optimizada predefinida), mas por razes pedaggicas.

Desta forma, com um lao *for* em *python*, o cdigo ficaria assim:

```python
def factorial(x):
  resultado=1
  for i in range(1,x+1):
    resultado*=i
  return resultado
```

Utilizando recurso o cdigo fica assim:

```python
def factorial(x):
  if x == 1:
    return 1
  else:
    return x * factorial(x-1)
```

A implementao  muito semelhante  definio que vimos acima. Vamos analisar o cdigo:
- O *if*  o caso base, sem ele a funo entraria em loop.
- O *else*  quem de fato faz toda mgica acontecer.

## **Calculando a Exponenciao**

Outro exemplo consiste na definio de exponenciao. Como sabemos, se $x\neq 0$, tem-se:

\begin{align*}
	x^n&=
	\begin{cases}
		1 & \mbox{ se } n=0,\\
		x.x^{n-1}& \mbox{ caso contrrio }.
	\end{cases}
\end{align*}

```python
def exp(x, n):
  if n == 0:
    return 1
  else:
    return x * exp(x,n-1)
exp(2,3)
```

## **Calculando o Mximo Divisor Comum**

Mais um exemplo, particularmente importante,  o denominado algoritmo de Euclides para o clculo eficiente do mximo divisor comum entre dois nmeros. A sua ideia bsica est no fato de `mdc(x,y) == mdc(y,x%y)`. O caso base  atingido quando algum dos valores  zero.

```python
def mdc(x, y):
  if x == 0 or y == 0:
    return x + y
  else:
    return mdc(y,x%y)
```

2. Desenvolver funes que realizem buscas em pastas. Ento, apresente:

    a) o espao ocupado por todos os arquivos:

```python
import os

def total_size(dir):
  size=0
  for name in os.listdir(dir):
    path = os.path.join(dir,name)
    if os.path.isdir(path):
      size+=total_size(path)
    else:
      size+=os.path.getsize(path)
  return size

total_size('sample_data')/1e6
```

    b) o caminho para o maior e menor arquivo:

```python
import os
import math

def larger_smaller(dir):
  larger={'path':None,'size':-math.inf}
  smaller={'path':None,'size':math.inf}
  for name in os.listdir(dir):
    path = os.path.join(dir,name)
    if os.path.isdir(path):
      larger_temp,smaller_temp=larger_smaller(path)
      if larger_temp['size']>larger['size']:
        larger=larger_temp
      if smaller_temp['size']<smaller['size']:
        smaller=smaller_temp
    else:
      aux={'path':path,'size':os.path.getsize(path)}
      if aux['size']>larger['size']:
        larger=aux
      if aux['size']<smaller['size']:
        smaller=aux
  return larger,smaller

larger_smaller('sample_data')
```

    c) uma lista com os tipos de arquivos:

```python
import os

def extensions(dir):
  alist=[]
  for name in os.listdir(dir):
    path = os.path.join(dir,name)
    if os.path.isdir(path):
      for ext in extensions(path):
        if ext not in alist:
          alist.append(ext)
    else:
      _,ext=os.path.splitext(path)
      if ext not in alist:
        alist.append(ext)
  return alist

extensions('sample_data')
```

    d) a quantidade de arquivos por tipo:

```python
import os

def count_extension(dir,extension):
  counter=0
  for name in os.listdir(dir):
    path=os.path.join(dir,name)
    if os.path.isdir(path):
      counter+=count_extension(path,extension)
    else:
      _,ext = os.path.splitext(path)
      if ext == extension:
        counter+=1
  return counter

dir = 'sample_data'
for extension in extensions(dir):
  count = count_extension(dir,extension)
  print(extension, count)
```

# **Glossrio**

- A programao recursiva assenta num princpio bsico da resoluo de problemas: reduzir passo a passo cada problema a outro(s) mais simples, at chegar a casos bsicos cuja soluo seja imediata.
- A recurso tira partido, portanto, da estrutura indutiva dos valores do domnio das funes que se pretendem calcular.
- A concepo de algoritmos  uma tarefa criativa, que exige muita prtica, mas que consiste essencialmente na boa compreenso do problema, na formulao mental da soluo, e finalmente na respectiva codificao.
