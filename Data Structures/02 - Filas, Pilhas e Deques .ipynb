# **Filas**

Uma fila (*queue*) é uma coleção de itens em que a inserção de novos itens acontece em uma extremidade, chamado de “fim” (*rear*), e a remoção de itens existente ocorre no outro extremo, comumente chamado de “início” (*front*). Um elemento é inserido no fim da fila e faz o seu caminho em direção ao início, esperando até aquele momento em que é o próximo elemento a ser removido. O item inserido mais recentemente na fila deve aguardar no final do coleção. O item que está na coleção há mais tempo está mais próximo do início. Este princípio de ordenação é às vezes chamado de FIFO, primeiro a entrar, primeiro a sair (first-in first-out). Também é conhecido como “primeiro a chegar, primeiro a ser servido” (first-come first-serverd).

## **Resumo**

- São estruturas lineares do tipo **FIFO**:
 - **Inserções** ocorrem no **final** da fila
 - **Exclusões** ocorrem no **início** da fila

## **Exemplos**:

- Fila do banco;
- Fila de impressão;
- Etc.

## Implementação

```
class Fila:
  
    # inicialização
    def __init__(self):
        self.__lista = []

    # tornar a fila iterável
    def __getitem__(self,posicao):
      return self.__lista[posicao]
    
    # comprimento
    def __len__(self):
      return len(self.__lista)

    # exibir
    def __repr__(self):
        return 'inicio -> ' + f'Fila: {self.__lista}' + ' <- fim'

    # inserir elemento
    def insercao(self, item):
        self.__lista.append(item) # adicionar um item ao final da lista

    # remover elemento
    def remocao(self):
        return self.__lista.pop(0) # remove o item do início da lista
```

# **Pilhas**

Uma pilha (*stack*) é uma coleção ordenada de itens onde a inserção de novos itens e a remoção de itens existentes sempre ocorrem na mesma extremidade. Esta extremidade é comumente chamada de topo. A extremidade oposta ao topo é conhecida como base. Os itens armazenados na pilha que estão mais perto da base são aqueles que estão na pilha a mais tempo. O último item inserido é aquele que está na posição em que pode ser removido primeiro. Este princípio de ordenação é as vezes chamado de LIFO, último a entrar primeiro a sair (*last in first out*). Os itens estão armazenados de acordo com o tempo de permanência na estrutura. Itens recentemente inseridos na estrutura estão mais próximos do topo, enquanto os itens mais antigos na estrutura estão perto da base.

## **Resumo**

- São estruturas lineares do tipo **LIFO**:
 - **Inserções** ocorrem no **topo** da pilha
 - **Exclusões** ocorrem no **topo** da pilha

## **Exemplos**:

- Pilha de pratos;
- Pilha de execução;
- Etc.

Ver: https://pt.wikipedia.org/wiki/Pilha_de_chamada

##**Implementação**

```
class Pilha:

    # inicializar
    def __init__(self):
        self.__lista = []

    # tornar a pilha iterável
    def __getitem__(self,posicao):
      return self.__lista[posicao]

    # comprimento
    def __len__(self):
      return len(self.__lista)
    
    # exibir
    def __repr__(self):
        return f'Pilha: {self.__lista}' + ' <- topo'

    # topo
    def topo(self):
      return self.__lista[len(self) - 1]

    # inserir
    def inserir(self, item):
        self.__lista.append(item) # adicionar um item ao final da lista
    
    # remover
    def remover(self):
        return self.__lista.pop() # remover o último item da lista
```

# **Deques**

## **Considerações**

É uma estrutura linear na qual os elementos podem ser **inseridos ou excluídos de qualquer** uma de suas **extremidades** (do início ou do fim). É conhecida por fila duplamente terminada, do inglês **D**ouble **E**nded **Que**ue.

## **Exemplos**

Sistemas distribuídos sempre necessitam que algum tipo de processamento seja mais rápido, por ser mais prioritário naquele momento, deixando outros tipos mais lentos ou em fila de espera, por não requerem tanta pressa.

ver: https://pt.wikipedia.org/wiki/Escalonamento_de_processos

## **Implementação**

```
class Deque:

    # inicializar
    def __init__(self):
        self.__lista = []

    # tornar o deque iterável
    def __getitem__(self,posicao):
      return self.__lista[posicao]

    # comprimento
    def __len__(self):
      return len(self.__lista)

    # exibir
    def __repr__(self):
        return 'traseira -> ' + f'Deque: {self.__lista}' + ' <- dianteira'
    
    # inserir na dianteira
    def inserir_dianteira(self, item):
        self.__lista.append(item) # adicionar um item ao final da lista
    
    # inserir na traseira
    def inserir_traseira(self, item):
        self.__lista.insert(0,item) # adicionar um item no início da lista

    # remover da dianteira
    def remover_dianteira(self):
        return self.__lista.pop() # removendo um item do final da lista

    # Remover da traseira
    def remover_traseira(self):
        return self.__lista.pop(0) # removendo um item do início da lista
```

## **Execução**

#### Inicializar um deque

```
deque = Deque()
```

#### Inserir processos no deque

```
import collections

Processo = collections.namedtuple('Processo',['nome', 'prioridade'])
```

```
deque.inserir_dianteira(Processo('p1',0))
deque.inserir_traseira(Processo('p2',1))
deque.inserir_dianteira(Processo('p3',0))
deque.inserir_traseira(Processo('p4',1))
```

#### Exibir processos no deque

```
print(deque)
```

#### Buscar processo no deque

```
Processo('p1',0) in deque
```

#### Contar processos no deque

```
len(deque)
```

#### Remover processos do deque

```
deque.remover_traseira()
```

```
deque.remover_dianteira()
```

```
print(deque)
```

## **Exercício**

**1.** Um problema interessante que pode ser facilmente resolvido usando a estrutura de dados deque (fila dupla) é o problema clássico do palíndromo. Um palíndromo é uma string que é lida da mesma forma do início para o fim e do fim para o início, por exemplo, radar, toot e madam são palídromos. Gostaríamos de construir um algoritmo para verificar se uma string é um palíndromo.

A solução para este problema usará um deque para armazenar os caracteres da string. Vamos processar a string da esquerda para a direita e inserir cada caractere no fim do deque. Neste ponto, o deque estará atuando agindo de maneira muito parecida com uma fila normal. No entanto, agora podemos fazer uso da a dupla funcionalidade do deque. O início da deque amazenará o primeiro caractere da string e o fim da deque manterá o último caractere.

image.png

Como podemos remover os ambos diretamente, podemos compará-los e continuar apenas se esses caracteres forem iguais. Se pudermos manter a correspondência entre os itens do início e do fim, ao final ficaremos sem caracteres na deque ou ficar com uma deque de tamanho 1, dependendo se o comprimento da string original era par ou ímpar.

```
def eh_palindromo(string):
  deque = Deque()
  for l in string:
    deque.inserir_dianteira(l)
  while len(deque) > 1:
    dianteira = deque.remover_dianteira()
    traseira = deque.remover_traseira()
    if dianteira != traseira:
      return False
  return True

print(eh_palindromo('tiago'))   
```
