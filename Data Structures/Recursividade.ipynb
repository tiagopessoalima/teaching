# Recursividade

A recursividade está intimamente ligada na matemática. Este conceito é bastante utilizado em desenvolvimento de sistemas, principalmente em algoritmos da disciplina de Estrutura de Dados. Pode ser definida como sendo um objeto que se define em termos de si próprio. Em outras palavras: uma função que chama a si própria até a sua resolução. Sua estratégia consiste em dividir a resolução de um problema grande em vários problemas pequenos.

## Três leis básicas da recursividade

1. O algoritmo recursivo deve ter um caso base, ou seja, um caso em que ele vai ser encerrado.

2. O algoritmo recursivo deve mudar o seu estado e se aproximar do caso base a cada chamada.

3. O algoritmo recursivo deve chamar a si mesmo.

## Exemplo

### Fatorial
É um número natural inteiro positivo, o qual é representado por $n!$ O fatorial de um número é calculado pela multiplicação desse número por todos os seus antecessores até chegar ao número $1$.

image.png

```python
def fatorial(n):
  fat = n
  for i in range(1,n):
    fat*=i
  return fat

print(fatorial(5))
```

```python
def fatorial(n):
  if n == 1:
    return 1
  else:
    return n * fatorial(n-1)
    
print(fatorial(5))
```

### Soma de uma lista de números

```python
def listsum(numList):
  theSum = 0
  for i in numList:
    theSum += i
  return theSum

print(listsum([1,3,5,7,9]))
```

```python
def listsum(numList):
  if len(numList) == 1:
    return numList[0]
  else:
    return numList[0] + listsum(numList[1:])

print(listsum([1,3,5,7,9]))
```
