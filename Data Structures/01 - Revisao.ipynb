## Considerações

* A palavra reservada **class** indica a definição de uma classe de nome Fracao. Como na definição de funções ou blocos em Python, o nome deve ser seguido por :

* Todos os métodos com nomes entre \_\_ são especiais por serem tratados de forma diferente dos demais métodos.

  * O método especial \_\_init\_\_(self) é chamado automaticamente sempre que um objeto da classe correspondente é criado. Tipicamente, esse método é utilizado para criar os atributos com seus valores iniciais e, por isso, esse método é conhecido como construtor da classe.

  * O método especial \_\_str\_\_(self) deve retornar uma string com a representação textual do objeto. Essa função é utilizada por outras funções do Python, em particular pela função print().

  * O método especial \_\_repr\_\_(self) o console interativo e o debugger chamam repr nos resultados das expressões avaliadas. Além disso, também é chamado quando um __ str __(self) personalizado não estiver disponível.

* Um dos princípios da linguagem Python é tornar as coisas explícitas. O self é a forma do Python para permitir que objetos possam usufruir de seus próprios (self ~ próprio) métodos e atributos.

## Operações com Frações

```
class Fracao:
  def __init__(self, n=0, d=1):
    self.num = n
    self.den = d
  def __repr__(self):
    return '%d/%d'%(self.num, self.den)
  def get(self):
    return self.num, self.den
  def put(self, n=0, d=1):
    self.num, self.den = n, d
  def mul(self, outra_fracao):
    n = self.num * outra_fracao.num
    d = self.den * outra_fracao.den
    return Fracao(n,d)
```

```
f1 = Fracao(2,3)
f2 = Fracao(3,4)
f3 = f1.mul(f2)
print(f3)
```

## **Glossário**

* **Atributo**:
Um dos itens nomeados de dados que compõem uma instância.

* **Classe**:
Um tipo de composto definido pelo usuário. Uma classe também pode ser pensada como um molde para os objetos que são instâncias da mesma.

* **Construtor**:
Cada classe tem uma “fábrica”, chamada pelo mesmo nome da classe, por fazer novas instâncias. Se a classe tem um método de inicialização, este método é usado para obter os atributos (ou seja, o estado) do novo objeto adequadamente configurado.

* **Instância**:
Um objeto cujo tipo é de alguma classe. Instância e objeto são usados como sinônimos.

* **Instanciar**:
Significa criar uma instância de uma classe e executar o seu método de inicialização.

* **Linguagem Orientada a Objetos**:
Uma linguagem que fornece recursos, como as classes definidas pelo usuário e herança, que facilitam a programação orientada a objetos.

* **Método**:
Uma função que é definida dentro de uma definição de classe e é chamado em instâncias dessa classe.

* **Método de Inicialização**:
Um método especial em Python, chamado __init__(), é chamado automaticamente para configurar um objeto recém-criado no seu estado inicial (padrão de fábrica).

* **Objeto**:
Um tipo de dados composto que é frequentemente usado para modelar uma coisa ou conceito do mundo real. Ele agrupa os dados e as operações que são relevantes para esse tipo de dados. Instância e objeto são usados como sinônimos.

* **Programação Orientada à Objetos**:
Um estilo poderoso de programação em que os dados e as operações que os manipulam são organizados em classes e métodos.

1. Altere a classe Fracao para que, quando o denominador for 1, ele imprima apenas o valor do numerador. Exemplo: para a chamada print(Fracao(3,1)) seja impresso o valor 3 e não 3/1.

```
class Fracao:
  def __init__(self, n=0, d=1):
    self.num = n
    self.den = d
  def __repr__(self):
    if self.den==1:
      return '%d'%(self.num)
    else:
      return '%d/%d'%(self.num, self.den)
  def get(self):
    return self.num, self.den
  def put(self, n=0, d=1):
    self.num, self.den = n, d
  def mul(self, outra_fracao):
    n = self.num * outra_fracao.num
    d = self.den * outra_fracao.den
    return Fracao(n,d)
```

2. Altere o construtor da classe Fracao para que, quando o numerador e o denominador possam ser reduzidos, a fração reduzida seja armazenada. Exemplo: para a chamada Fracao(12,20) o valor de self.num e sel.den devem ser, respectivamente, 3 e 5. Dica: crie uma função ou, se preferir, um método, com mdc que recebe 2 números inteiros a e b e retorne o máximo divisor comum entre a e b

```
class Fracao:

  def mdc(self, a, b):
    while b !=0:
        a, b = b, a % b
    return a

  def __init__(self, n=0, d=1):
    v = self.mdc(n,d)
    self.num = n/v
    self.den = d/v

  def __repr__(self):
    if self.den==1:
      return str(self.num)
    else:
      return str(self.num) +'/'+ str(self.den)
```

```
f = Fracao(12,20)
print(f)
```

**Multiplicação e Divisão:** Entre a multiplicação e divisão de frações, a multiplicação é a operação mais simples, pois não é necessário encontrar o denominador em comum para os cálculos. Sendo assim, na multiplicação de frações basta multiplicar numerador com numerador (termos da parte superior), e denominador com denominador (termos da parte inferior).

Entenda melhor com os exemplos a seguir:

2/3 . 5/3 = 2.5 / 3.3 = 10/9

6/4 . 7/9 = 6.7 / 42/36 = 7/6

8/9 . 6/15 = 8.6 / 9.15 = 48/135 = 16/45

A divisão de frações também serve para a simplificação dos termos fracionários (numeradores e denominadores). Para tais operações, é preciso conservar a primeira fração e multiplicar pelo inverso da segunda.

Ou seja, os numeradores assumem os lugares dos denominadores, e vice-versa. Vejamos as seguintes frações:

2/3 : 5/6 = 2/3 . 6/5 = 12 / 15 = 4/5

4/5 : 7/4 = 4/5 . 4/7 = 16/ 35

4/9 : 6 = 4/9. 1/6 = 4/54 = 2/27

```
class Fracao:

  def mdc(self, a, b):
    while b !=0:
        a, b = b, a % b
    return a

  def __init__(self, n=0, d=1):
    v = self.mdc(n,d)
    self.num = n/v
    self.den = d/v

  def __sub__(self, outra_fracao):
    n = self.num - outra_fracao.num
    d = self.den - outra_fracao.den
    return Fracao(n,d)

  def __repr__(self):
    if self.den==1:
      return str(self.num)
    else:
      return str(self.num) +'/'+ str(self.den)
```

```
f1 = Fracao(2,3)
f2 = Fracao(3,4)
f3 = f1 - f2
print(f3)
```
