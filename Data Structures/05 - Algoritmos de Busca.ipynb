### **Busca Sequencial**

Percorrer a lista comparando cada elemento **alist[i]** com **x**. Caso o valor seja encontrado retorna **True** e, caso **alist[i] > x** retorna **False**. Tamb√©m retorna **False** se n√£o encontrar o elemento.

#### **Implementa√ß√£o**

```
def busca_sequencial(alist,x):

  passos = 0

  for elemento in alist:

    passos += 1

    if elemento == x:

      return True, passos

    elif elemento > x:

      return False, passos

  return False, passos
```
