### **Bubble Sort**

Realiza m√∫ltiplas passagem por uma lista. Ele compara itens adjacentes e troca aqueles que est√£o fora de ordem. Cada passagem pela lista coloca o pr√≥ximo maior valor na sua posi√ß√£o correta. A figura abaixo ilustra a primeira passada do algoritmo.

No come√ßo da segunda passagem, o maior valor agora est√° ordenado. Como cada passagem coloca o pr√≥ximo maior valor encontrado no lugar certo, o n√∫mero total de passagens necess√°rias para ordenar um vetor de tamanho/comprimento $n$ ser√° $ùëõ‚àí1$.

####**Swap**

A troca de dois elementos numa lista requer o uso de uma vari√°vel para armazenamento tempor√°rio, conforme trecho de c√≥digo abaixo:

```
alist = [3,2,1] # [2,3,1]

temp = alist[0] # temp -> 3
alist[0] = alist[1] # alist -> [2,2,1]
alist[1] = temp # alist -> [2,3,1]

print(alist)
```

Por√©m, em Python, √© poss√≠vel realizar atribui√ß√µes simult√¢neas, conforme exemplo abaixo:

```
alist = [3,2,1]

alist[0], alist[1] = alist[1], alist[0]

print(alist)
```

####**Implementa√ß√£o**

```
def bubbleSort(alist):

  n = len(alist)

  for p in range(n-1): # passagens do algoritmo

    for i in range(0,n-1-p): # itens adjacentes

      if alist[i] > alist[i+1]:

        alist[i], alist[i+1] = alist[i+1], alist[i]

    print('passagem', p+1, alist)
```

```
#alist = [54,26,93,17,77,31,44,55,20]
alist = [54,26,103,17,27,31,4,55,20]

bubbleSort(alist)
```

### **Selection Sort (Ordena√ß√£o Por Sele√ß√£o)**

Melhora o bubble sort, ao realizar apenas uma troca a cada passagem pela lista. Para conseguir isso, uma ordena√ß√£o por sele√ß√£o procura pelo valor mais alto enquanto faz uma passagem e, depois de complet√°-la, coloca-o na posi√ß√£o correta. Assim como o bubble sort, depois da primeira passagem, o maior item sempre est√° na posi√ß√£o correta. Depois da segunda passagem, o pr√≥ximo maior item tamb√©m vai para a posi√ß√£o adequada. O processo continua dessa forma e demanda $ùëõ‚àí1$ passagens para ordenar $ùëõ$ itens, j√° que o item final dever√° estar no lugar certo depois da ($ùëõ‚àí1$)-√©sima passagem. A figura abaixo ilustra o processo de ordena√ß√£o completo.

image.png

Anima√ß√£o: https://visualgo.net/en/sorting

####**Implementa√ß√£o**

```
def selectionSort(alist):

  n = len(alist)

  for slot in range(n-1,0,-1):

    i_max=0

    for i in range(1,slot+1):

      if alist[i]>alist[i_max]:

        i_max = i

    alist[slot], alist[i_max] = alist[i_max], alist[slot]

    print('slot', slot, alist)
```

#### **Execu√ß√£o**

```
alist = [54,26,93,17,77,31,44,55,20]
selectionSort(alist)
```

#### **An√°lise**

Voc√™ pode ver que a ordena√ß√£o por sele√ß√£o faz o mesmo n√∫mero de compara√ß√µes que o bubble sort. Portanto, tamb√©m √© $ùëÇ(ùëõ¬≤)$. Contudo, devido √† redu√ß√£o do n√∫mero de trocas, a ordena√ß√£o por sele√ß√£o tipicamente executa mais r√°pido em an√°lises de desempenho. ### **Insertion Sort (Ordena√ß√£o por Inser√ß√£o)**

Embora tamb√©m seja $ùëÇ(ùëõ¬≤)$, funciona de uma forma ligeiramente diferente. Ele sempre mant√©m uma sublista ordenada nas posi√ß√µes inferiores da lista. Cada novo item √© ‚Äúinserido‚Äù na sublista anterior de modo que a sublista ordenada fique com um item a mais. A figura abaixo mostra o processo de ordena√ß√£o por inser√ß√£o. Itens sombreados representam as sublistas ordenadas conforme cada passagem feita pelo algoritmo.

Anima√ß√£o: https://visualgo.net/en/sorting

Come√ßamos pressupondo que uma lista com um √∫nico item (posi√ß√£o 0) est√° ordenada. A cada passagem, ou seja, do item $1$ a $ùëõ‚àí1$, o item atual √© comparado com os que j√° est√£o na sublista ordenada. Conforme vamos caminhando de tr√°s pra frente na sublista j√° ordenada, movemos os itens que s√£o maiores para a direita. Quando encontramos um item menor ou chegamos ao fim da sublista, o item atual √© inserido.

A figura abaixo mostra a quinta passagem em detalhe. Nesse ponto do algoritmo, h√° uma sublista com cinco itens: 17, 26, 54, 77 e 93. Queremos inserir 31 nesse conjunto j√° ordenado. A primeira compara√ß√£o faz com que o item 93 v√° para a direita. Os itens 77 e 54 tamb√©m s√£o deslocados para frente. Quando o item 26 √© encontrado, o processo de deslocamento para e o 31 √© colocado na lacuna aberta. Temos agora uma sublista ordenada com seis itens.

image.png

####**Implementa√ß√£o**

```
def insertionSort(alist):

  n = len(alist)

  for i in range(1,n):

    value, position = alist[i], i

    while position>0 and alist[position-1]>value:

      alist[position]=alist[position-1]

      position -= 1

    alist[position]=value

    print(alist)
```

####**Execu√ß√£o**

```
alist = [54,26,93,17,77,31,44,55,20]
insertionSort(alist)
```

####**An√°lise**

O n√∫mero m√°ximo de compara√ß√µes para uma ordena√ß√£o por inser√ß√£o √© a soma dos primeiros $ùëõ‚àí1$ inteiros. De novo, isso √© $ùëÇ(ùëõ¬≤)$. Contudo, no melhor caso, apenas uma compara√ß√£o √© necess√°ria a cada passagem. Esse seria o caso de uma lista j√° ordenada. Uma observa√ß√£o importante sobre a diferen√ßa entre deslocamento e troca: em geral, a opera√ß√£o de deslocamento requer aproximadamente um ter√ßo de processamento de uma troca, j√° que apenas uma atribui√ß√£o √© feita. ### **Shell Sort**

√â o mais eficiente algoritmo de ordena√ß√£o dentre os de complexidade quadr√°tica $ùëÇ(ùëõ¬≤)$. √â um refinamento do m√©todo *Insertion Sort*, melhorando a ordena√ß√£o ao quebrar a lista original em sublistas, as quais s√£o ordenadas usando a t√©cnica de ordena√ß√£o por inser√ß√£o. Isso pode ser visto nas figuras abaixo, numa lista com  nove itens. Se usarmos um incremento de tr√™s, ser√£o tr√™s sublistas, cada uma sendo submetida √† ordena√ß√£o por inser√ß√£o.

A figura abaixo mostra uma ordena√ß√£o por inser√ß√£o final usando um incremento de um. Observa-se que ao realizar as ordena√ß√µes de sublistas anteriores, o n√∫mero total de opera√ß√µes de deslocalmento necess√°rias para colocar a lista na sua ordem final √© reduzido. Nesse caso, precisa-se de apenas mais quatro deslocamentos para completar o processo.



image.png

####**Implementa√ß√£o**

```
def gapInsertionSort(alist,inicio,gap):

  for i in range(inicio+gap,len(alist),gap):

    valor_atual, posicao = alist[i], i

    while posicao>=gap and alist[posicao-gap]>valor_atual:

      alist[posicao]=alist[posicao-gap]

      posicao = posicao-gap

    alist[posicao]=valor_atual
```

```
def shellSort(alist):

  tamanho = len(alist)//2

  while tamanho > 0:

    for posicao_inicial in range(tamanho):

      gapInsertionSort(alist,posicao_inicial,tamanho)

    print("Depois do incremento de tamanho de", tamanho, "a lista √©",alist)

    tamanho = tamanho // 2
```

####**Execu√ß√£o**

```
alist = [54,26,93,17,77,31,44,55,20]
shellSort(alist)
print(alist)
```

####**An√°lise**

Pode-se pensar que o *Shell Sort* n√£o tem como ser melhor que o *Insertion Sort*, j√° que ele a utiliza na lista toda como um √∫ltimo passo. Acontece que essa ordena√ß√£o por inser√ß√£o final n√£o precisa realizar muitos deslocamentos, j√° que a lista foi pr√©-ordenada por ordena√ß√µes por inser√ß√£o incrementais anteriores. Em outras palavras, cada passagem produz uma lista que est√° ‚Äúmais ordenada‚Äù que a anterior. Isso faz com que a passagem final seja muito eficiente. Embora uma an√°lise geral do *Shell Sort* esteja al√©m do escopo deste documento, pode-se dizer que ele tende a ficar entre $ùëÇ(ùëõ)$ e $ùëÇ(ùëõ¬≤)$. Alterando o incremento, por exemplo, usando $2^k‚àí1$ (1, 3, 7, 15, 31, e assim por diante), o *Shell Sort* pode executar a $O(n^{3/2})$.

### **Merge Sort (Ordena√ß√£o por Jun√ß√£o)**


Algoritmo recursivo que divide uma lista continuamente pela metade. Se ela estiver vazia ou possuir um √∫nico item, est√° ordenada por defini√ß√£o (caso base). Se possuir mais de um item, divide-se a lista e invoca-se recursivamente um *Merge Sort* em ambas metades. Assim que as metades estiverem ordenadas, a opera√ß√£o fundamental, chamada de intercala√ß√£o, √© realizada. Intercalar √© o processo de pegar duas listas menores ordenadas e combin√°-las de modo a formar uma lista nova, √∫nica e ordenada. A figura abaixo mostra uma lista sendo dividida pelo *Merge Sort*.

####**Implementa√ß√£o**

```
def merge(alist, lefthalf, righthalf):

    i_left, i_right, i_list = 0, 0, 0

    while i_left < len(lefthalf) and i_right < len(righthalf):

      if lefthalf[i_left] < righthalf[i_right]:

        alist[i_list]=lefthalf[i_left]

        i_left+=1

      else:

        alist[i_list]=righthalf[i_right]

        i_right+=1

      i_list+=1

    while i_left < len(lefthalf):

      alist[i_list]=lefthalf[i_left]

      i_left+=1

      i_list+=1

    while i_right < len(righthalf):

      alist[i_list]=righthalf[i_right]

      i_right+=1

      i_list+=1

    print("Merging ",alist)
```

```
def mergeSort(alist):

  print("Splitting ",alist)

  if len(alist) > 1:

    mid = len(alist) // 2

    lefthalf = alist[:mid]

    righthalf = alist[mid:]

    mergeSort(lefthalf)

    mergeSort(righthalf)

    merge(alist, lefthalf, righthalf)
```

####**Execu√ß√£o**

```
alist = [54,26,93,17,77,31,44,55,20]
mergeSort(alist)
print(alist)
```

### **Quick Sort**

Primeiro seleciona um valor, chamado de **piv√¥**. O papel do **piv√¥** √© ajudar na divis√£o da lista. A posi√ß√£o √† qual o **piv√¥** pertence de fato na lista ordenada, conhecida como ponto de divis√£o, ser√° usada para quebrar a lista em chamadas subsequentes do *quick sort*. A figura abaixo mostra que o primeiro elemento funcionar√° como o primeiro **piv√¥**.

image.png

O processo de parti√ß√£o ocorrer√° em seguida. Ele ir√° encontrar o ponto de divis√£o e, ao mesmo tempo, mover√° os itens para o lado apropriado da lista, isto √©, maior ou menor que o valor do **piv√¥**. O particionamento come√ßa localizando dois marcadores de posi√ß√£o: *leftmark* e *rightmark* ‚Äî no come√ßo e no fim dos demais itens da lista. O objetivo do processo de particionamento √© mover os itens que est√£o do lado errado do **piv√¥** para o lado certo, convergindo para o ponto de divis√£o, conforme mostra a figura abaixo.

image.png

O algoritmo come√ßa incrementando o *leftmark* at√© encontrar um valor que seja maior que o valor do **piv√¥**. Ent√£o, decrementa-se o *rightmark* at√© que se encontre um valor que seja menor que o valor do **piv√¥**. Nesse instante, descobre-se dois itens que est√£o fora de ordem em rela√ß√£o ao eventual ponto de divis√£o. Para o exemplo, isso ocorre com o $93$ e o $20$. Agora √© poss√≠vel trocar esses itens e repetir o processo.

No instante em que *rightmark* se tornar menor que o *leftmark* significa que foi encontrado o ponto de divis√£o. O valor do **piv√¥** pode ser trocado com o que estiver no ponto de divis√£o e o valor do **piv√¥** estar√° agora em seu devido lugar. A lista pode agora ser dividida no ponto de divis√£o e o *quicksort* pode ser chamado recursivamente sobre as duas metades, conforme figura abaixo.


image.png

####**Implementa√ß√£o**

```
def partition(alist,first,last):

  pivotvalue = alist[first]

  leftmark = first+1

  rightmark = last

  while True:

    while leftmark <= rightmark and alist[leftmark] <= pivotvalue:

      leftmark = leftmark + 1

    while alist[rightmark] >= pivotvalue and rightmark >= leftmark:

      rightmark = rightmark - 1

    if rightmark < leftmark:

      break

    else:

      alist[leftmark], alist[rightmark] = alist[rightmark], alist[leftmark]

  alist[first], alist[rightmark] = alist[rightmark], alist[first]

  return rightmark
```

```
def quickSort(alist,first=0,last=None):

  if last == None:

    last = len(alist)-1

  if first<last:

    splitpoint = partition(alist,first,last)

    quickSort(alist,first,splitpoint-1) # left

    quickSort(alist,splitpoint+1,last) # right
```

####**Execu√ß√£o**

```
alist = [54,26,93,17,77,31,44,55,20]
quickSort(alist)
print(alist)
```

####**An√°lise**

Para uma lista de tamanho $n$, se a parti√ß√£o sempre ocorre no meio da lista, haver√° de novo $logn$ divis√µes. Contudo, n√£o h√° necessidade de mem√≥ria adicional como no caso do *merge sort*. Para encontrar o ponto de divis√£o, cada um dos $n$ itens precisa ser comparado com o valor do **piv√¥**. O resultado √© $nlogn$. No pior caso, os pontos de divis√£o podem n√£o estar no meio. Na verdade, eles podem tender muito √† esquerda ou √† direita, criando uma separa√ß√£o desigual. Nesse caso, ordenar uma lista de $n$ itens poderia dividir a lista em $0$ itens de um lado $n‚àí1$ do outro. Ordenar uma lista de $n-1$, por sua vez, resultaria na divis√£o de uma lista de tamanho $0$ de um lado e $n‚àí2$ do outro, e assim por diante. O resultado disso √© uma ordena√ß√£o da ordem $O(n¬≤)$ com o custo tipicamente requerido por chamadas recursivas.
